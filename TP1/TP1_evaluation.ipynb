```python
from num2words import num2words
import re
X2=[re.sub(r'\b(\d+)(st|nd|rd|th)\b', lambda m: num2words(int(m.group(1)), to='ordinal'), text) for text in X]#goup(1) je prends seulement les chiffres
X2 = [re.sub(r'\b\d+\b', lambda m: num2words(int(m.group())), text) for text in X2]
print(X2[0])
print("-----------------------------------------------------------------------------------")
print(X2[40])
```

![ ](https://img.shields.io/badge/drop_stop_words%20%20-blue)

```python
X3 = []
for text in X2:
    text_sp = nlp(text)
    tokens = [t for t in text_sp if not t.is_stop]
    X3.append(tokens)
print (X3[0])   
```

![ ](https://img.shields.io/badge/drop_punctuations%20%20-blue)

```python
X4 = []
for text in X3:
    tokens = []
    for token in text:
        if token.text not in punctuations :
            tokens.append(token.text)#.text!
    X4.append(tokens)
print(X4[0])
```

![ ](https://img.shields.io/badge/Process%20%20-orange)

```python
X_processed = []
for text in X4:
    text_pro = nlp(' '.join(text))
    lemmas = []
    for token in text_pro:
        lemmas.append(token.lemma_)
    X_processed.append(lemmas)

print(X_processed[0])
```
